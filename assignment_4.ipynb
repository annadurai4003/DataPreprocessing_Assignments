```python
################FINDING OUTLIERS##################
import numpy as np
for col_name in loan_data.select_dtypes(include=np.number).columns[:-1]:
    print(col_name)
    q1 = loan_data[col_name].quantile(0.25)
    q3 = loan_data[col_name].quantile(0.75)
    iqr = q3 - q1
    
    low = q1-1.5*iqr
    high = q3+1.5*iqr
    print("Outliers = ",loan_data.loc[(loan_data[col_name] < low) | (loan_data[col_name] > high), col_name])
```
